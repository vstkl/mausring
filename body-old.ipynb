```python
from jupyterscad import view
from  solid2 import * 

# dimensions 
debug_matrix = [0,0,20]
null_matrix = [0,0,0]
debug = True
render = False
set_global_fn(200)

base_r = 14
base_d = 4
base_w = 1
gap = 2
tolerance = 1

off = 0.5

chip_esp_a=4
chip_esp_h=1
chip_esp_b=6
chip_bca_a=2

tolerance_matrix = [0,0,-tolerance]

inner_r = base_r 
inner_d = base_d
inner_w = base_w 

material_min = 0.5

outer_r = base_r + base_w + gap
outer_d = base_d
outer_w = base_w +material_min


mag_count = 8
mag_x,mag_y,mag_z = [1,1,base_d]
mag_a = 1
mag_b = base_d
mag_out_a = mag_a+off*tolerance
mag_out_b = mag_b+2

magnets = []
magnets_out = []
for magnet in range(0, mag_count):
    magnets += cube([mag_a,mag_a,mag_b]).translate([outer_r-mag_a-material_min,(base_d-mag_z)]).rotate([0,0,((magnet*(360/mag_count)+22))])
    magnets_out += cube([mag_out_a,mag_out_a,mag_out_b]).translate([outer_r-mag_a-material_min,0,(base_d-mag_z)]).translate(-(tolerance*off)/2,-(tolerance*off)/2,-1).rotate([0,0,((magnet*(360/mag_count)+22))])
    

inner_base = cylinder(inner_d,inner_r,inner_r) 
inner_diff = cylinder(inner_d + 2*tolerance,inner_r - inner_w,inner_r - inner_w ).translate(tolerance_matrix)

chip_esp = cube(chip_esp_a,chip_esp_a,chip_esp_h).translate(-chip_esp_a/2,0,inner_r-inner_w).rotate(90,0,0).translate(0,-off/2,0)
chip_esp_out = cube(chip_esp_b,chip_esp_b,chip_esp_h+off).translate(-chip_esp_b/2,0,inner_r-inner_w).rotate(90,0,0).translate(0,0,-1)
chip_bca = cube(chip_bca_a,chip_bca_a,1).translate(-chip_esp_a/2,0,0).rotate(90,0,0)

inner = inner_base - inner_diff - chip_esp_out
outer_base = cylinder(outer_d,outer_r,outer_r) 
outer_diff = cylinder(outer_d + 2*tolerance,outer_r - outer_w,outer_r - outer_w).translate(tolerance_matrix)

outer = outer_base - outer_diff - magnets_out


inner = inner.color([0,0,0.7])
outer = outer.color([0,0,0.9])
magnets = magnets.color([0.9,0,0])
to_render = [inner, outer,  chip_esp, magnets]
to_debug = []

obj = [] 

for rendered in to_render:
    obj += rendered
    
for debugged in to_debug:
    obj += debugged.translate(debug_matrix)
    
if render:
    obj.save_as_scad("body.scad")
    obj.save_as_stl("body.stl")

view(obj) 
```
